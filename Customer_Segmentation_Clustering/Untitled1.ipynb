```python
import pandas as pd

df_offers = pd.read_excel("./WineKMC.xlsx", sheet_name=0)
df_offers.columns = ["offer_id", "campaign", "varietal", "min_qty", "discount", "origin", "past_peak"]
df_offers.head()
```

```python
df_transactions = pd.read_excel("./WineKMC.xlsx", sheet_name=1)
df_transactions.columns = ["customer_name", "offer_id"]
df_transactions['n'] = 1
df_transactions.head()
```

```python
df = pd.merge(df_offers, df_transactions)
matrix = df.pivot_table(index=['customer_name'], columns=['offer_id'], values='n')
matrix = matrix.fillna(0).reset_index()
matrix.head()
```

```python
import numpy as np
# Select all rows and exclude the "customer_name" column from table2
x_cols = matrix.iloc[:,1:]
# Convert the DF into an array using the array function
x_cols = np.array(x_cols)
x_cols
```

```python
from sklearn.cluster import KMeans

# Number of clusters
kmeans = KMeans(n_clusters=3)
# Fitting the input data
kmeans = kmeans.fit(x_cols)
# Getting the cluster labels
labels = kmeans.predict(x_cols)
# Centroid values
centroids = kmeans.cluster_centers_
print(centroids)
print(len(centroids))
```
